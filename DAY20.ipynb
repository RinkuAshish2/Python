```python
# Create scatter plot
sns.scatterplot(data=cars, x='HP', y='MPG')
plt.title('Scatter Plot of Horsepower vs. Miles Per Gallon')
plt.xlabel('Horsepower (HP)')
plt.ylabel('Miles Per Gallon (MPG)')
plt.show()
```

```python
# Calculate correlation
correlation = cars['HP'].corr(cars['MPG'])
print(f'Correlation between HP and MPG: {correlation}')
```
